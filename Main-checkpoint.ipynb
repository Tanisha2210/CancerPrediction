This pie chart serves as a fundamental first step in our analysis, providing a clear visual summary of the class distribution within our dataset. It answers a critical question: Are our two classes, Acute Lymphoblastic Leukemia (ALL) and Acute Myeloid Leukemia (AML), represented equally?

Here, we observe a distribution of approximately 73% for ALL and 27% for AML. This tells us that the dataset is imbalanced—there are significantly more samples from ALL patients than from AML patients.

Why this is important at a graduate level:
This imbalance has significant implications for our modeling strategy. Many machine learning algorithms are sensitive to class imbalance because they are designed to maximize overall accuracy. If left unaddressed, our models could become biased towards the majority class (ALL), learning to predict it well at the expense of misclassifying the minority class (AML). In a clinical context, failing to correctly identify the rarer AML cases would be a critical error. Therefore, this initial observation immediately informs our subsequent steps, signaling the need for techniques like stratified sampling during cross-validation, using evaluation metrics beyond simple accuracy (like F1-score, precision, and recall), and potentially employing advanced methods such as SMOTE (Synthetic Minority Over-sampling Technique) or using class weights in our models to ensure both cancer types are treated with appropriate importance.

Why this is important at a graduate level:
In information theory, variance is often a proxy for information content. A gene with near-zero variance (i.e., it's expressed at the same constant level in every patient, regardless of their condition) offers no useful information for classification. ```python

```python
def preprocess_data(X, y, pca_variance_threshold=0.85):
    """Preprocess the data: scaling and PCA"""
    
    print(f"\nDATA PREPROCESSING")
    print("="*50)
    
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                        random_state=42, stratify=y)
    
    print(f"Train set: {X_train.shape[0]} samples")
    print(f"Test set: {X_test.shape[0]} samples")
    
    # Standardize features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print(f"Features standardized")
    
    # Apply PCA
    pca = PCA()
    X_train_pca_full = pca.fit_transform(X_train_scaled)
    
    # Find number of components for desired variance
    cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
    n_components = np.argmax(cumsum_variance >= pca_variance_threshold) + 1
    
    print(f"PCA Analysis:")
    print(f"- Components for {pca_variance_threshold*100}% variance: {n_components}")
    print(f"- Actual variance retained: {cumsum_variance[n_components-1]:.3f}")
    
    # Apply PCA with selected components
    pca_selected = PCA(n_components=n_components)
    X_train_pca = pca_selected.fit_transform(X_train_scaled)
    X_test_pca = pca_selected.transform(X_test_scaled)
    
    print(f" PCA applied: {X.shape[1]} → {n_components} features")
    
    # Visualize PCA results
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Explained variance plot
    ax1.plot(range(1, min(21, len(pca.explained_variance_ratio_) + 1)), 
             pca.explained_variance_ratio_[:20], 'bo-', markersize=8)
    ax1.axhline(y=0.05, color='r', linestyle='--', alpha=0.7)
    ax1.set_title('PCA Explained Variance Ratio', fontweight='bold')
    ax1.set_xlabel('Principal Component')
    ax1.set_ylabel('Explained Variance Ratio')
    ax1.grid(True, alpha=0.3)
    
    # Cumulative variance plot
    ax2.plot(range(1, len(cumsum_variance) + 1), cumsum_variance, 'ro-', markersize=6)
    ax2.axhline(y=pca_variance_threshold, color='g', linestyle='--', alpha=0.7, 
                label=f'{pca_variance_threshold*100}% threshold')
    ax2.axvline(x=n_components, color='g', linestyle='--', alpha=0.7)
    ax2.set_title('PCA Cumulative Explained Variance', fontweight='bold')
    ax2.set_xlabel('Number of Components')
    ax2.set_ylabel('Cumulative Explained Variance')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return (X_train, X_test, y_train, y_test, 
            X_train_scaled, X_test_scaled,
            X_train_pca, X_test_pca, 
            scaler, pca_selected)

# Preprocess the data
(X_train, X_test, y_train, y_test, 
 X_train_scaled, X_test_scaled,
 X_train_pca, X_test_pca, 
 scaler, pca_selected) = preprocess_data(X, y)
```

```python
# Baseline Classifiers
import pandas as pd
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, accuracy_score, f1_score, precision_score, recall_score

def build_robust_baseline_classifiers(X_train, y_train):
    """Builds and evaluates baseline classifiers using cross-validation for robust performance estimates."""
    print(f"\nROBUST BASELINE ANALYSIS (via Cross-Validation)")
    print("="*60)
    
    # Define the models to evaluate
    classifiers = {
        'Dummy (Most Frequent)': DummyClassifier(strategy='most_frequent', random_state=42),
        'Dummy (Stratified)': DummyClassifier(strategy='stratified', random_state=42),
        'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')
    }
    
    # Define the metrics to compute. We handle the zero_division issue here.
    scoring = {
        'accuracy': make_scorer(accuracy_score),
        'f1': make_scorer(f1_score, zero_division=0),
        'precision': make_scorer(precision_score, zero_division=0),
        'recall': make_scorer(recall_score, zero_division=0)
    }
    
    results = {}
    
    # Perform 5-fold cross-validation for each classifier
    for name, model in classifiers.items():
        print(f"Running CV for: {name}...")
        cv_results = cross_validate(model, X_train, y_train, cv=5, scoring=scoring)
        
        # Store the mean of the results
        results[name] = {
            'Accuracy': cv_results['test_accuracy'].mean(),
            'F1 Score': cv_results['test_f1'].mean(),
            'Precision': cv_results['test_precision'].mean(),
            'Recall': cv_results['test_recall'].mean(),
            'Fit Time (s)': cv_results['fit_time'].mean()
        }
    
    # Display results in a DataFrame
    baseline_df = pd.DataFrame(results).T
    print("\nRobust Baseline Classifier Performance (5-fold CV on Training Data):")
    print(baseline_df.round(3))
    
    return baseline_df

# Usage example:
baseline_df = build_robust_baseline_classifiers(X_train_pca, y_train)
```

NAIVE BAYES CLASSIFIER WITH CROSS-VALIDATION

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import (accuracy_score, f1_score, precision_score, 
                           recall_score, roc_auc_score, classification_report, 
                           confusion_matrix)
import matplotlib.pyplot as plt
import seaborn as sns

def load_genetic_data(filepath):
    """Load and prepare the genetic dataset"""
    print("Loading genetic dataset...")
    
    try:
        # Load the CSV file
        data = pd.read_csv(filepath)
        print(f"Dataset loaded successfully!")
        print(f"Shape: {data.shape}")
        print(f"Columns: {list(data.columns[:5])}... (showing first 5)")
        
        # Separate features and target
        X = data.drop('Target', axis=1)
        y = data['Target']
        
        print(f"Features shape: {X.shape}")
        print(f"Target distribution: {y.value_counts().to_dict()}")
        
        return X, y
        
    except FileNotFoundError:
        print(f"Error: File '{filepath}' not found!")
        return None, None
    except Exception as e:
        print(f"Error loading data: {e}")
        return None, None

def preprocess_data(X, y, test_size=0.2, apply_pca=True, n_components=0.95):
    """Preprocess the genetic data"""
    print(f"\nPreprocessing data...")
    
    # Check for missing values
    missing_values = X.isnull().sum().sum()
    if missing_values > 0:
        print(f"Warning: {missing_values} missing values found. Filling with mean.")
        X = X.fillna(X.mean())
    
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )
    
    print(f"Train set: {X_train.shape[0]} samples")
    print(f"Test set: {X_test.shape[0]} samples")
    
    # Standardize features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    if apply_pca:
        # Apply PCA for dimensionality reduction
        pca = PCA(n_components=n_components, random_state=42)
        X_train_pca = pca.fit_transform(X_train_scaled)
        X_test_pca = pca.transform(X_test_scaled)
        
        n_components_used = X_train_pca.shape[1]
        explained_variance = pca.explained_variance_ratio_.sum()
        
        print(f"PCA applied: {X.shape[1]} -> {n_components_used} components")
        print(f"Explained variance: {explained_variance:.3f}")
        
        return (X_train_pca, X_test_pca, y_train, y_test, 
                scaler, pca, n_components_used)
    else:
        return (X_train_scaled, X_test_scaled, y_train, y_test, 
                scaler, None, X.shape[1])

def implement_naive_bayes(X_train, X_test, y_train, y_test, cv_folds=5):
    """
    Implement Naive Bayes classifier with proper error handling
    """
    print(f"\nNAIVE BAYES CLASSIFIER")
    print("="*50)
    
    # Validate input data
    if X_train is None or len(X_train) == 0:
        print("Error: Training data is empty or None!")
        return None, None, None, None, None
    
    print(f"Training on {X_train.shape[0]} samples with {X_train.shape[1]} features")
    print(f"Test set: {X_test.shape[0]} samples")
    print(f"Class distribution in training: {np.bincount(y_train)}")
    
    try:
        # Initialize classifier
        nb_classifier = GaussianNB()
        
        # Cross-validation
        cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
        
        # Perform cross-validation
        scores = cross_validate(
            nb_classifier, X_train, y_train, cv=cv,
            scoring=['accuracy', 'f1'], return_train_score=False
        )
        
        cv_scores = scores['test_accuracy']
        cv_f1_scores = scores['test_f1']
        
        # Display CV results
        print(f"\n Cross-Validation Results (Accuracy):")
        print(f"- Mean: {cv_scores.mean():.3f} (+/- {cv_scores.std() * 2:.3f})")
        print(f"- Individual folds: {[f'{score:.3f}' for score in cv_scores]}")
        
        print(f"\n Cross-Validation Results (F1-Score):")
        print(f"- Mean: {cv_f1_scores.mean():.3f} (+/- {cv_f1_scores.std() * 2:.3f})")
        
        # Train on full training set
        nb_classifier.fit(X_train, y_train)
        
        # Make predictions
        y_pred_nb = nb_classifier.predict(X_test)
        
        # Calculate probabilities for binary classification
        if len(np.unique(y_test)) == 2:
            y_pred_proba_nb = nb_classifier.predict_proba(X_test)[:, 1]
            auc = roc_auc_score(y_test, y_pred_proba_nb)
        else:
            y_pred_proba = None
            auc = None
        
        # Display test results
        print(f"\nTest Set Performance:")
        print(f"- Accuracy: {accuracy_score(y_test, y_pred_nb):.3f}")
        print(f"- F1-Score: {f1_score(y_test, y_pred_nb, average='weighted'):.3f}")
        print(f"- Precision: {precision_score(y_test, y_pred_nb, average='weighted'):.3f}")
        print(f"- Recall: {recall_score(y_test, y_pred_nb, average='weighted'):.3f}")
        
        if auc is not None:
            print(f"- AUC: {auc:.3f}")
        
        # Detailed classification report
        print(f"\nDetailed Classification Report:")
        print(classification_report(y_test, y_pred_nb))
        
        # Confusion matrix
        print(f"\nConfusion Matrix:")
        cm = confusion_matrix(y_test, y_pred_nb)
        print(cm)
        
        return nb_classifier, y_pred_nb, y_pred_proba_nb, cv_scores, cv_f1_scores
        
    except Exception as e:
        print(f"Error in Naive Bayes implementation: {e}")
        return None, None, None, None, None

def plot_results(cv_scores, cv_f1_scores, y_test, y_pred_nb, y_pred_proba_nb=None):
    """Plot training results"""
    plt.figure(figsize=(15, 5))
    
    # Plot 1: CV scores
    plt.subplot(1, 3, 1)
    plt.boxplot([cv_scores, cv_f1_scores], labels=['Accuracy', 'F1-Score'])
    plt.title('Cross-Validation Scores')
    plt.ylabel('Score')
    
    # Plot 2: Confusion Matrix
    plt.subplot(1, 3, 2)
    cm = confusion_matrix(y_test, y_pred_nb)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # Plot 3: ROC curve (if binary classification)
    if y_pred_proba_nb is not None:
        from sklearn.metrics import roc_curve
        plt.subplot(1, 3, 3)
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba_nb)
        auc = roc_auc_score(y_test, y_pred_proba_nb)
        plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.3f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
        plt.legend()
    
    plt.tight_layout()
    plt.show()

def main():
    """Main pipeline execution"""
    print("GENETIC DATA MACHINE LEARNING PIPELINE")
    print("="*60)
    
    # Step 1: Load data
    filepath = "../expanded_genetic_data.csv"  # Update this path if needed
    X, y = load_genetic_data(filepath)
    
    if X is None:
        print(" Failed to load data. Please check the file path.")
        return
    
    # Step 2: Preprocess data
    results = preprocess_data(X, y, test_size=0.2, apply_pca=True)
    X_train, X_test, y_train, y_test, scaler, pca, n_components = results
    
    # Step 3: Train Naive Bayes
    nb_results = implement_naive_bayes(X_train, X_test, y_train, y_test)
    nb_model, y_pred_nb, y_pred_proba_nb, nb_cv_scores, nb_cv_f1 = nb_results
    
    if nb_model is not None:
        # Step 4: Plot results
        print("\nGenerating visualization...")
        plot_results(nb_cv_scores, nb_cv_f1, y_test, y_pred_nb, y_pred_proba_nb)
        
        # Step 5: Model summary
        print(f"\nMODEL TRAINING COMPLETED!")
        print(f"- Final model trained on {X_train.shape[0]} samples")
        print(f"- Features used: {n_components}")
        print(f"- Test accuracy: {accuracy_score(y_test, y_pred_nb):.3f}")
        
        # Save model if needed
        # import joblib
        # joblib.dump(nb_model, 'naive_bayes_genetic_model.pkl')
        # print("Model saved as 'naive_bayes_genetic_model.pkl'")
    
    else:
        print("Model training failed!")

# Run the pipeline
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB


try:
    full_df = pd.read_csv('../expanded_genetic_data.csv')
    informative_features = full_df.columns[1:51].tolist()
    X = full_df[informative_features]
    y = full_df['Target']
except FileNotFoundError:
    print("Error: 'train_data.csv' not found. Creating dummy data.")
    X = pd.DataFrame(np.random.rand(37, 50))
    y = pd.Series(np.random.randint(0, 2, 37))

# Preprocessing to get PCA-transformed data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
pca = PCA(n_components=0.85) # Using the same 85% variance threshold
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print("Data is preprocessed. PCA resulted in {} features.".format(X_train_pca.shape[1]))
print("-" * 50)

print("\nSTEP 1: TRAINING PHASE")
print("The model learns statistics for each class (ALL=0, AML=1).")

# Separate the training data by class
X0 = X_train_pca[y_train == 0]
X1 = X_train_pca[y_train == 1]

# Calculate the statistics (mean, variance, and priors)
# These learned parameters are the entire "model"
model_stats = {
    'class_0': {
        'prior': len(X0) / len(X_train_pca),
        'mean': X0.mean(axis=0),
        'var': X0.var(axis=0)
    },
    'class_1': {
        'prior': len(X1) / len(X_train_pca),
        'mean': X1.mean(axis=0),
        'var': X1.var(axis=0)
    }
}

print(f"\nPrior probability for ALL (Class 0): {model_stats['class_0']['prior']:.3f}")
print(f"Prior probability for AML (Class 1): {model_stats['class_1']['prior']:.3f}")
print(f"\nLearned Mean for Class 0 (first 3 features): \n{model_stats['class_0']['mean'][:3]}")
print(f"\nLearned Variance for Class 0 (first 3 features): \n{model_stats['class_0']['var'][:3]}")


print("\n\n STEP 2: PREDICTION PHASE")
print("Predicting the class for a single sample from the test set.")

# Select the first sample from the test set as our example
sample_to_predict = X_test_pca[0]
print(f"\nSample's feature values (first 3): {sample_to_predict[:3]}")

def gaussian_pdf(x, mean, var):
    """Calculate the probability density for a value given a Gaussian distribution."""
    # Add a small epsilon to variance to prevent division by zero
    epsilon = 1e-9
    numerator = np.exp(-((x - mean) ** 2) / (2 * (var + epsilon)))
    denominator = np.sqrt(2 * np.pi * (var + epsilon))
    return numerator


# For Class 0 (ALL)
log_prior_0 = np.log(model_stats['class_0']['prior'])
# Sum the log probabilities for each feature (this is the "naive" independence assumption)
log_likelihood_0 = np.sum(np.log(gaussian_pdf(sample_to_predict,
                                             model_stats['class_0']['mean'],
                                             model_stats['class_0']['var'])))
log_posterior_0 = log_prior_0 + log_likelihood_0

# For Class 1 (AML)
log_prior_1 = np.log(model_stats['class_1']['prior'])
log_likelihood_1 = np.sum(np.log(gaussian_pdf(sample_to_predict,
                                             model_stats['class_1']['mean'],
                                             model_stats['class_1']['var'])))
log_posterior_1 = log_prior_1 + log_likelihood_1


print("\n--- Calculation for Class 0 (ALL) ---")
print(f"Log Prior: {log_prior_0:.3f}")
print(f"Log Likelihood: {log_likelihood_0:.3f}")
print(f"Final Log Posterior Score: {log_posterior_0:.3f}")

print("\n--- Calculation for Class 1 (AML) ---")
print(f"Log Prior: {log_prior_1:.3f}")
print(f"Log Likelihood: {log_likelihood_1:.3f}")
print(f"Final Log Posterior Score: {log_posterior_1:.3f}")

# The prediction is the class with the higher log posterior score
manual_prediction = 0 if log_posterior_0 > log_posterior_1 else 1
print(f"\nResult: The Log Posterior for Class {manual_prediction} is higher.")
print(f"Manual Prediction: {manual_prediction} ({'ALL' if manual_prediction == 0 else 'AML'})")


print("\n\n STEP 3: VERIFICATION")
print("Comparing our manual result with scikit-learn's result.")

# Train the real scikit-learn model
sklearn_nb = GaussianNB()
sklearn_nb.fit(X_train_pca, y_train)

# Predict the same sample
sklearn_prediction = sklearn_nb.predict(sample_to_predict.reshape(1, -1))[0]

print(f"\nScikit-learn's Prediction: {sklearn_prediction} ({'ALL' if sklearn_prediction == 0 else 'AML'})")

# Check if our manual calculation matches
if manual_prediction == sklearn_prediction:
    print("\nSuccess! Our manual calculation matches the scikit-learn model's prediction.")
else:
    print("\nOur manual calculation does not match the scikit-learn model.")
```

LOGISTIC REGRESSION WITH HYPERPARAMETER TUNING

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

def implement_logistic_pipeline(X, y, cv_folds=5, pca_variance_threshold=0.85, tune=False):
    print("\nPIPELINE: SCALING + PCA + LOGISTIC REGRESSION")
    print("="*60)

    # Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                        random_state=42, stratify=y)
    print(f"Train: {X_train.shape}, Test: {X_test.shape}")

    # PCA: determine n_components from variance threshold
    pca_temp = PCA().fit(StandardScaler().fit_transform(X_train))
    cumsum_variance = np.cumsum(pca_temp.explained_variance_ratio_)
    n_components = np.argmax(cumsum_variance >= pca_variance_threshold) + 1
    print(f"PCA: {pca_variance_threshold*100:.1f}% variance => {n_components} components")

    # Pipeline setup
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('pca', PCA(n_components=n_components)),
        ('clf', LogisticRegression(max_iter=1000, solver='liblinear', random_state=42))
    ])

    if tune:
        param_grid = {
            'clf__C': [0.001, 0.01, 0.1, 1, 10],
            'clf__penalty': ['l1', 'l2']
        }
        cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
        grid = GridSearchCV(pipeline, param_grid, cv=cv, scoring='f1', n_jobs=-1, verbose=1)
        grid.fit(X_train, y_train)
        model = grid.best_estimator_
        print(f"Best Params: {grid.best_params_}")
        print(f"Best CV F1: {grid.best_score_:.3f}")
    else:
        model = pipeline.fit(X_train, y_train)

    # Cross-validation
    cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
    cv_acc = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    cv_f1 = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1')

    print("\nCross-Validation Results:")
    print(f"- Accuracy: {cv_acc.mean():.3f} (+/- {cv_acc.std()*2:.3f})")
    print(f"- F1 Score: {cv_f1.mean():.3f} (+/- {cv_f1.std()*2:.3f})")

    y_pred = model.predict(X_test)
    # y_pred_proba_lr = lr.predict_proba(X_test_pca)[:, 1]
    y_prob = model.predict_proba(X_test)[:, 1]
    

    print("\nTest Set Performance:")
    print(f"- Accuracy: {accuracy_score(y_test, y_pred):.3f}")
    print(f"- F1 Score: {f1_score(y_test, y_pred):.3f}")
    print(f"- Precision: {precision_score(y_test, y_pred):.3f}")
    print(f"- Recall: {recall_score(y_test, y_pred):.3f}")
    print(f"- AUC: {roc_auc_score(y_test, y_prob):.3f}")

    return model, y_pred, y_prob, cv_acc, cv_f1
```

```python
model, y_pred, y_prob, cv_acc, cv_f1 = implement_logistic_pipeline(X, y, tune=True)
```

```python
def create_final_visualizations():
    """Create comprehensive visualizations for model comparison"""
    
    print(f"\nCREATING FINAL VISUALIZATIONS")
    print("="*50)
    
    # Prepare model predictions and probabilities
    models_data = {
        # 'Naive Bayes': (y_pred_nb, y_pred_proba_nb),
        'Logistic Regression': (y_pred_lr, y_pred_proba_lr),
        'Random Forest': (y_pred_rf, y_pred_proba_rf),
        'XGBoost': (y_pred_xgb, y_pred_proba_xgb),
        'Neural Network': (y_pred_nn, y_pred_proba_nn)
    }
    
    # Create subplot figure
    fig = plt.figure(figsize=(20, 15))
    
    # ROC Curves Comparison
    ax1 = plt.subplot(2, 3, 1)
    colors = ['blue', 'red', 'green', 'orange', 'purple']
    
    for i, (model_name, (_, proba)) in enumerate(models_data.items()):
        fpr, tpr, _ = roc_curve(y_test, proba)
        auc_score = roc_auc_score(y_test, proba)
        ax1.plot(fpr, tpr, color=colors[i], label=f'{model_name} (AUC = {auc_score:.3f})')
    
    ax1.plot([0, 1], [0, 1], 'k--', alpha=0.5)
    ax1.set_title('ROC Curves - All Models', fontweight='bold', fontsize=14)
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Performance Metrics Comparison
    ax2 = plt.subplot(2, 3, 2)
    metrics = ['Accuracy', 'F1-Score', 'Precision', 'Recall', 'AUC']
    model_names = list(final_results.index)
    
    x = np.arange(len(metrics))
    width = 0.15
    
    for i, model in enumerate(model_names[:5]):  # Top 5 models
        values = [final_results.loc[model, metric] for metric in metrics]
        ax2.bar(x + i*width, values, width, label=model, alpha=0.8)
    
    ax2.set_title('Performance Metrics Comparison', fontweight='bold', fontsize=14)
    ax2.set_xlabel('Metrics')
    ax2.set_ylabel('Score')
    ax2.set_xticks(x + width * 2)
    ax2.set_xticklabels(metrics, rotation=45)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Confusion Matrix for Best Model (highest F1-score)
    best_f1_model = final_results['F1-Score'].idxmax()
    if best_f1_model == 'Naive Bayes':
        best_pred = y_pred_nb
    elif best_f1_model == 'Logistic Regression':
        best_pred = y_pred_lr
    elif best_f1_model == 'Random Forest':
        best_pred = y_pred_rf
    elif best_f1_model == 'XGBoost':
        best_pred = y_pred_xgb
    else:
        best_pred = y_pred_nn
    
    ax3 = plt.subplot(2, 3, 3)
    cm_best = confusion_matrix(y_test, best_pred)
    sns.heatmap(cm_best, annot=True, fmt='d', cmap='Blues', ax=ax3,
                xticklabels=['ALL', 'AML'], yticklabels=['ALL', 'AML'])
    ax3.set_title(f'Confusion Matrix - {best_f1_model}\n(Best F1-Score)', fontweight='bold', fontsize=14)
    ax3.set_xlabel('Predicted')
    ax3.set_ylabel('Actual')
    
    # Model Accuracy Ranking
    ax4 = plt.subplot(2, 3, 4)
    accuracy_sorted = final_results['Accuracy'].sort_values(ascending=True)
    colors_bar = plt.cm.viridis(np.linspace(0, 1, len(accuracy_sorted)))
    
    bars = ax4.barh(range(len(accuracy_sorted)), accuracy_sorted.values, color=colors_bar)
    ax4.set_yticks(range(len(accuracy_sorted)))
    ax4.set_yticklabels(accuracy_sorted.index)
    ax4.set_title('Model Accuracy Ranking', fontweight='bold', fontsize=14)
    ax4.set_xlabel('Accuracy')
    ax4.grid(True, alpha=0.3)
    
    # Add value labels on bars
    for i, (bar, value) in enumerate(zip(bars, accuracy_sorted.values)):
        ax4.text(value + 0.01, i, f'{value:.3f}', va='center')
    
    # F1-Score vs AUC Scatter Plot
    ax5 = plt.subplot(2, 3, 5)
    f1_scores = final_results['F1-Score']
    auc_scores = final_results['AUC']
    
    scatter = ax5.scatter(f1_scores, auc_scores, c=range(len(f1_scores)), 
                         cmap='viridis', s=100, alpha=0.7)
    
    for i, model in enumerate(final_results.index):
        ax5.annotate(model, (f1_scores[i], auc_scores[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=10)
    
    ax5.set_title('F1-Score vs AUC', fontweight='bold', fontsize=14)
    ax5.set_xlabel('F1-Score')
    ax5.set_ylabel('AUC')
    ax5.grid(True, alpha=0.3)
    
    # Model Complexity vs Performance
    ax6 = plt.subplot(2, 3, 6)
    # Simplified complexity ranking (subjective)
    complexity_ranking = {
        'Naive Bayes': 1,
        'Logistic Regression': 2,
        'Linear SVM': 3,
        'RBF SVM': 4,
        'Random Forest': 5,
        'XGBoost': 6,
        'XGBoost + PCA': 6.5,
        'Neural Network': 7
    }
    
    complexity_scores = [complexity_ranking.get(model, 5) for model in final_results.index]
    performance_scores = final_results['F1-Score'].values
    
    ax6.scatter(complexity_scores, performance_scores, s=100, alpha=0.7, c='coral')
    for i, model in enumerate(final_results.index):
        ax6.annotate(model, (complexity_scores[i], performance_scores[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=9, rotation=15)
    
    ax6.set_title('Model Complexity vs F1-Score', fontweight='bold', fontsize=14)
    ax6.set_xlabel('Model Complexity (Subjective Scale)')
    ax6.set_ylabel('F1-Score')
    ax6.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Create final visualizations
create_final_visualizations()
```
